```python
import ast
ast.literal_eval('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')

## was in dictionary form and we evaluated the reliability of this string
```

* From the data in some columns, the dictionary named name pull data

```python
def donusum(genres):
        L = []
        for i in ast.literal_eval(genres):
            L.append(i["name"])
        return L   
```

```python
movies_copy["genres"]=movies_copy["genres"].apply(donusum)
```

```python
movies_copy["keywords"]=movies_copy["keywords"].apply(donusum)
```

```python
def donusum1(genres):
    L = []
    counter =0
    for i in ast.literal_eval(genres):
            
        if counter !=3:
                L.append(i["name"])
                counter +=1
        else:
                break
    return L   
    
## With this function, you can extract the names in the cast and crew data.
## We will take 3 of them
```

```python
movies_copy["cast"]=movies_copy["cast"].apply(donusum1)
```

* In the Crew section, it is enough to see only the director, the other team is not very important and increases the data size.

```python
def yonetmen(obje):
    L=[]
    for i in ast.literal_eval(obje):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L    
```

```python
movies_copy["crew"]=movies_copy["crew"].apply(yonetmen)
```

* we inserted commas between words in the text to make it easier to process and feed them to machine learning models

```python
movies_copy["overview"]=movies_copy["overview"].apply(lambda x:x.split())
```

* We want the model to be more accurate and meaningful in order to avoid confusion in cases where there are the same name but different surnames.

```python
movies_copy.genres=movies_copy.genres.apply(lambda x:[i.replace(" ","")for i in x])
movies_copy.keywords=movies_copy.keywords.apply(lambda x:[i.replace(" ","")for i in x])
movies_copy.cast=movies_copy.cast.apply(lambda x:[i.replace(" ","")for i in x])
movies_copy.crew=movies_copy.crew.apply(lambda x:[i.replace(" ","")for i in x])
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

## With this library we find the root forms of words, for example
## input : running  output:run
```

```python
def stem(metın):
    lıste = []
    
    for i in metın.split():
            lıste.append(ps.stem(i))
            
    return " ".join(lıste)     
```

## Let's Create a Function That Recommends 5 Movies

```python
def recommender(movies):
    movie_index = new_df[new_df["title"]==movies].index[0]
    distances = benzerlik_metriği[movie_index]
    movies_LİST = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]  # Here we told him to make 5 movie suggestions. 
    
    for i in movies_LİST:
        print(new_df.iloc[i[0]].title) ## Get the index of the angles in the movies list and show where they correspond in the title column.
    
    
    return

```

```python
recommender("Batman")
```
